# Data Cleaning

```python
# Drop columns that aren't needed for visualization, or have so many missing value

df.drop(["id", "NAME", "host id", "host name", "lat", "long", "house_rules", "license", 'country', 'country code', 'last review', 'reviews per month'], axis=1, inplace=True)
df
```

```python
temp_df = df[['neighbourhood group', 'price']]
temp_df
```

```python
temp_df=temp_df.dropna()
```

```python
def remove_sign(price):
    return int(str(price).replace('$','').replace(' ','').replace(',',''))
```

```python
temp_df['price'].apply(remove_sign)
```
